```python
# importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn provides some example datasets. For example, on the [installation walkthrough](https://seaborn.pydata.org/installing.html), there is a penguin dataset:

```python
df = sns.load_dataset("penguins")
sns.pairplot(df, hue="species")
```

To set the overall appearance of all plots in the program, run this command:

```python
# Apply the default theme
sns.set_theme()
```

Note that if you re-run the previous cell, the graph would appear different. 

```python
# Load another example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

Since we recently went over linear regression, I wanted to include the built-in linear regression visualisation along with the confidence margin:

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

Here is one of Michael Waskom's *(Seaborn's creator)* favorite features of the library: the [violin plot](https://en.wikipedia.org/wiki/Violin_plot). After listening to his lecture at the CUNY Graduate Center colloquium, I found that plot very intuitive and was interested in playing with it in Python.

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

Finally, I really liked the joint plot feature for the data visualization purposes. Here's the smokers vs non-smokers tip example:

```python
tips = sns.load_dataset("tips")
sns.jointplot(data=tips, x="total_bill", y="tip", hue="smoker")
```
